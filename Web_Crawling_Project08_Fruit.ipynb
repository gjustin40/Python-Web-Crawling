```python
# ImageNet에서 받기
import urllib
import os

def get_url_list(url):
    data = urllib.request.urlopen(url).read()
    url_list = data.decode('utf-8').split('\r\n')
    
    return url_list

# image url list
orange_url = 'http://www.image-net.org/api/text/imagenet.synset.geturls?wnid=n07747607'
strawberry_url = 'http://www.image-net.org/api/text/imagenet.synset.geturls?wnid=n07745940'
grape_url = 'http://www.image-net.org/api/text/imagenet.synset.geturls?wnid=n07758680'
banana_url = 'http://www.image-net.org/api/text/imagenet.synset.geturls?wnid=n07753592'
apple_url = 'http://www.image-net.org/api/text/imagenet.synset.geturls?wnid=n07739125'
pineapple_url = 'http://www.image-net.org/api/text/imagenet.synset.geturls?wnid=n07753275'
watermelon_url = 'http://www.image-net.org/api/text/imagenet.synset.geturls?wnid=n07756951'

fruit_url = {
    'watermelon': watermelon_url,    
}

for name, f_url in fruit_url.items():
    print(name, f_url)
    fruit_url_list = get_url_list(f_url)
    
    for i, url in enumerate(fruit_url_list):
        try:
            urllib.request.urlretrieve(url, '{}/{}{}.jpg'.format(name, name, i))
        except:
            print(i)
```
